```python
import openai
import os
import json

from dotenv import load_dotenv

load_dotenv()

openai.api_key = os.environ["OPENAI_API_KEY"]

model = "gpt-3.5-turbo-0613"
```

```python
'''
Elimina tildes
'''
def clean(word):
    word = word.replace('á', 'a')
    word = word.replace('é', 'e')
    word = word.replace('í', 'i')
    word = word.replace('ó', 'o')
    word = word.replace('ú', 'u')
    word = word.replace(' ', '')
    return word
```

```python
def addMesage(trio):

    messages = [
        {
        "role": "system",
        "content": """
        """
        },
        {
        "role": "user",
        "content": """
    
        Un frase entre 20 y 50 palabras, describiendo una escena usando estos tres conceptos: """ + trio + """, la respuesta
        debe seguir si o si (añade el símbolo #) el siguiente formato: "(frase en español) # (frase en ingles)"
        para las frases en español remueve tildes, ñ, y ".
        """
        }
    ]

    result = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0.7,
        max_tokens=100,
    )

    return result["choices"][0]["message"]["content"]

conceptos= ['Antropoceno',
            'Antropocentrismo',
            'Convivencia entre humanos y máquinas',
            'Desarrollo humano',
            'Desastre ecológico',
            'Ecología',
            'Ecosistemas',
            'Ética de la Inteligencia Artificial'
            'Extractivismo',
            'Inteligencia Artificial',
            'Modernización digital',
            'Mundo común',
            'Pérdida de biodiversidad',
            'Recursos naturales',
            'Regulación tecnológica']
```

```python
conceptos

trios = []
for i in range(len(conceptos)):
    for j in range(i+1, len(conceptos)):
        for k in range(j+1, len(conceptos)):
            trios.append(conceptos[i] + ", " + conceptos[j] + ", " + conceptos[k])
```

```python
import time

futureJSON = {"prompts_styles": [], "prompts_concepts": {}}

times = 0

for i in range(len(conceptos)):
    for j in range(i+1, len(conceptos)):
        for k in range(j+1, len(conceptos)):

            times += 1
            print(times, " de 455")

            trioGPT = conceptos[i] + ", " + conceptos[j] + ", " + conceptos[k]
            trioJSON = clean(conceptos[i]) + ", " + clean(conceptos[j]) + ", " + clean(conceptos[k])

            dict = {}

            dict["esp_prompts"] = []
            dict["eng_prompts"] = []

            for _ in range(5):

                try:
                    prompt = addMesage(trioGPT)
                    espanol = prompt[:prompt.index("#")]
                    ingles = prompt[prompt.index("#")+1:]

                    
                    dict["esp_prompts"].append(espanol.strip())
                    dict["eng_prompts"].append(ingles.strip())

                except:
                    print("Error: " + str(i+1) + "/455, " + str(j+1) + "/5")
                    print(conceptos)
                    print(prompt)
                    # wait 60 secs
                    print("Retrying...")
                    time.sleep(60)

            print(dict)

            
            futureJSON["prompts_concepts"][trioJSON] = dict



```

```python
with open('data.json', 'w') as outfile:
    json.dump(futureJSON, outfile)
```
